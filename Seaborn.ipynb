## 1. Distribution Plots
Let's discuss some plots that allow us to **visualize the distribution of a data set**. Thes plots are:
* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

## distplot
The distplot shows the distribution of a **univariate** set of observations.

```python
sns.distplot(tips['total_bill'])
```

To remove the kde layer and just have the histogram we set kde attribute to false as:

```python
sns.distplot(tips['total_bill'], kde=False, bins=20)
```

## jointplot
jointplot() allows us to basicaly match up two displots for **bivariate data**. With our choice of what **kind** parameter to compare with:
* scatter
* reg //regression
* resid
* kde
* hex

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') #kind is default to scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## pairplot
pairplot will plot pairwise relationship across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns)

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## rugplot
rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are building block of a KDE plot:

```python
sns.rugplot(tips['total_bill'])
```

## kdeplot
kdeplots are **KERNAL DENSITY ESTIMATION** plots. These KDE plots replace every single observation with a Guassian(Normal) distribution centered around that value.

```python
sns.kdeplot(tips['total_bill'])
```

# 2. Categorical Plots
Now let's see different types of categorical plots which we can plot using seaborn. Categorical plots used to plot **Categorical Data**. There are a few main plot types for this:
* barplot
* countplot
* boxplot //box and whisker plot
* stripplot
* violinplot
* swarmplot
* factorplot
Let's go through examples of each

```python
   tips.head()
```

## barplot and counplot
These very similar plots allow us to get aggregate data of a categorical feature in our data. **barplot** is a general plot that allows us to aggregate the categorical data based off some function, by default the **mean**:

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

here default estimator is **mean** and this indicates that avg. total_bill of males is greater than avg. total_bill of females.

We can change the change the estimator object to our own function, that converts a vector to scaler as:

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

## countplot
This is essentially the same as barplot except the estimator is explicitly counting the number of occurences(count), and which is why we only pass the x value:

```python
sns.countplot(x='sex',data=tips)
```

and this indicates the no. of males in the dataset is greater than no. of females.

## boxplot and violinplot
boxplots and violinplots are used to show the distribution of categorical data. A box plot shows the distribution of quantitative data in a way that facilitates comparisons bewteen variables or across levels of a categorical variable. The **box** shows the **quartiles** of the dataset while the **whiskers** extend to show the rest of the distribution, except for points that are determined to be **outliers** using a method that is a fucntion of the inter-quartile range.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='coolwarm')
```

```python
sns.boxplot(data=tips,orient='h')
```

## violinplot
A violinplot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one(or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernal density estimation of the underlying distribution.

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,palette='Set1')
```

## stripplot and swarmplot
The **stripplot** will draw a scatterplot where one variable is categorical. A stripplot cna be drawn on its own, but it is also a good complement to a box or violin plot in cases where we want to show all observations along with some representation of the underlying distribution.

The **swarmplot** is similar to stripplot, but the points are adjusted(only along the categorical axis) so that they don't overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations(both in terms of ability to show all the points and in terms of the computation needed to arange them).

```python
sns.stripplot(x='day',y='total_bill', data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips,palette='Set1')
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips,split=True,palette='Set1')
```

```python
#removing above warnings
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips,dodge=True,size=4)
```

## Combining Categorical Plots

```python
sns.violinplot(x='tip',y='day',data=tips,palette='rainbow')
sns.swarmplot(x='tip',y='day',data=tips,color='white',size=3)
```

## factorplot
factorplot is the most general form of categorical plot. It can take in a **kind** parameter to adjust the plot type

```python
sns.factorplot(x='sex',y='tip',data=tips,kind='bar')
```

```python
#working on above warning
sns.catplot(x='sex',y='tip',data=tips,kind='box')
```

# 3. Matrix Plots
Matrix plots allow us to plot data as color-encoded matrices and can also be used to indicate clusters within the data( it becomes more clear in Machine Learning that how to formally cluster data).
Let's explore Seaborn's heatmap and clustermap:

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights.head()
```

## Heatmap
In order for a heatmap to work properly, our data should already be in a matrix form, and the heatmap function basically just colors it for us.

```python
#Matrix form for correlation data
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(),annot=True)
```

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
```

```python
## converting flights dataset to matric form
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
fp = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(fp, cmap ='pink')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

## clustermap
The clustermap uses hierarchical clustering to produce a clustered version of the heatmap. 

```python
sns.clustermap(fp,cmap='pink')
```

Notice how the years and months ar no longer in order, instead they ar grouped by similarity in value(passenge count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months).

```python
# More options to get the information a little cleaner like noraml normalization
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

# 4. Grids
Grids are general types of plots that allow us to map plot types to rows and columns of a grid, this helps us to create similar plots separated by features.

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## PairGrid
PairGrid is a subplot for plotting pairwise relationships in a datasets.

```python
#Just the Grid
sns.PairGrid(iris)
```

```python
#Then we map the grid
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
#Map to Upper, Lower and Diagonal individually
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## pairplot
pairplot is a simpler version of PairGrid (we'll use quite often)

```python
sns.pairplot(iris)
```

## Facet Grid
FacetGrid is the general way to create grids of plots based off a feature

```python
tips.head()
```

```python
g = sns.FacetGrid(tips,col='time', row='smoker')
g = g.map(plt.hist,'total_bill')
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
g = g.map(plt.scatter,'total_bill','tip').add_legend()  #Notice how the arguments come after plt.scatter call
```

## JointGrid
JointGrid is the general version for jointplot() type grids.

```python
g = sns.JointGrid(x='total_bill',y='tip',data=tips)
```

```python
g = g.plot(sns.regplot, sns.distplot)
```

# 5. Regression Plots
Seaborn has many built-in capabilities fo regression plots, however **Regression** is a topic of Machine Learning. So here we won't really discuss it but we will only cover the **lmplot()** function for now.

**lmplot()** allows us to display linear models, but it convieniently allows us to split up those plots based off of features, as well as coloring the hue based off of feature.

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

## Working with Markers
lmplot kwargs get passed through to **regplot** which is more general form of lmplot(). regplot has scatter_kws parameter that gets passed to plt.scatter. So we want to set the s parameter in that dictionary, which corresponds(a bit confusingly) to the squared markersize. In otherwords we end up passing a dictionary with the base matplotlib arguments, in this case, s for size of a scatter plot. In general, we probably won't remember this off the top of our head, but instead reference the documentation.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid
We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time') #similar of FacetGrid , here these are bit automatic
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

## Aspect and Size
Seaborn figures can have their size and aspect ration adjusted with the **size** and **aspect** parameters.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=5)
```

# Style and Color of Plots
We have seen few times how to control figure asthetics in seaborn, but let's now go over it formally.

## Styles
We can set particular styles:

```python
sns.set_style('dark')  #default is 'white'
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

## Spine Removal

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

## Size and Aspect
We can use matplotlib's ***plt.figure(figsize=(width,height)*** to change the size of the most seaborn plots.
We can control the size and aspect ratio of most seaborn gird plots by passing in parameters: size and aspect

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',height=2,aspect=4,data=tips)  #size changed to height
```

## Scale and Context
The **set_context** allows us to override default parameters.

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```
